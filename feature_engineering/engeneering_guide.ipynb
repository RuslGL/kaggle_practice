# Imports, settings, common data

```python
import pandas as pd
import numpy as np
import warnings

from sklearn.datasets import fetch_california_housing

warnings.filterwarnings('ignore')
```

```python
data = fetch_california_housing(as_frame=True)
df =  data.data
df.columns = ['med_inc', 'house_age', 'ave_rooms', 'ave_bedrms', 'population', 'ave_ocup',
       'latitude', 'longitude']
short_data = df.head(100)

target = data.target
```

# Features types

[CONTENT](#CONTENT)

## Raw features
Directly taken from the original dataset 

eg age, city, price

## Derived (handcrafted)
Created through mathematical or logical transformations

eg log(price), ratio = revenue / cost 

## Aggregated

Computed statistics over groups or time

eg avg_transaction_value, weekly_sum

## Encoded / Vectorized
Converted from categorical, textual, or visual data 

eg one-hot, TF-IDF, label encoding

## Learned (neural)
Extracted using pretrained models or embeddings

BERT text vectors, CNN image embeddings
